```python
import pandas as pd

df = pd.read_csv("berlin_expenses.csv")
df
```

```python
df.head()
```

```python
df.info()
```

```python
df["date"] = pd.to_datetime(df["date"])

```

```python
df.info()
```

```python
df["month"] = df["date"].dt.to_period("M")
df.head()
```

```python
monthly_expenses = df.groupby("month")["amount"].sum()
monthly_expenses
```

```python
import matplotlib.pyplot as plt

monthly_expenses.plot(kind="bar")
plt.title("Monthly Expenses in Berlin (2024)")
plt.xlabel("Month")
plt.ylabel("Total Expense (€)")
plt.tight_layout()
plt.show()
```
