# Berlin Cost of Living – Data Cleaning & Analysis

## Dataset Description
This dataset contains personal monthly expenses in Berlin for 2024.
It includes:
- Date of expense
- Expense category
- Amount in EUR

## Data Cleaning Steps
- Converted date column to datetime format
- Checked for missing values
- Ensured correct data types

## Exploratory Data Analysis
- Calculated total monthly expenses
- Visualized monthly spending trends


```python
# Ay bilgisi çıkaralım
df["month"] = df["date"].dt.to_period("M")

# Aylık toplam harcama
monthly_expenses = df.groupby("month")["amount"].sum()

monthly_expenses
```

```python
import matplotlib.pyplot as plt

monthly_expenses.plot(kind="bar")

plt.title("Monthly Expenses in Berlin (2024)")
plt.xlabel("Month")
plt.ylabel("Total Expense (EUR)")
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

monthly_expenses.plot(kind="bar")

plt.title("Monthly Expenses in Berlin (2024)")
plt.xlabel("Month")
plt.ylabel("Total Expense (EUR)")
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# Kategori bazlı toplam harcama
category_expenses = df.groupby("category")["amount"].sum()

category_expenses
```

```python
import matplotlib.pyplot as plt

category_expenses.plot(kind="bar")

plt.title("Total Expenses by Category (2024)")
plt.xlabel("Category")
plt.ylabel("Total Expense (EUR)")
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Aylık toplam harcama zaten vardı ama garanti olsun
monthly_expenses = df.groupby("month")["amount"].sum()

# Line plot
plt.figure()
monthly_expenses.plot(kind="line", marker="o")

plt.title("Monthly Total Expenses (2024)")
plt.xlabel("Month")
plt.ylabel("Total Expense (EUR)")
plt.grid(True)
plt.tight_layout()

plt.show()
```
